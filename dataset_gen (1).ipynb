```python
import pandas as pd
import random
from faker import Faker
import numpy as np

# Initialize Faker
fake = Faker()

# Define the number of records
num_records = 100000

# Generate synthetic EV fleet data
data = {
    "Vehicle ID": [f"EV{str(i).zfill(4)}" for i in range(1, num_records + 1)],
    "Driver ID": [f"DR{str(i).zfill(4)}" for i in range(1, num_records + 1)],
    "Maintenance Cost ($)": [round(random.uniform(100, 2000), 2) for _ in range(num_records)],
    "Vehicle Status": [random.choice(["Active", "Under Maintenance", "Inactive"]) for _ in range(num_records)],
    "Battery Capacity (kWh)": [random.randint(50, 150) for _ in range(num_records)],
    "Pincode": [fake.postcode() for _ in range(num_records)],
    "Top Speed (km/h)": [random.randint(100, 200) for _ in range(num_records)],
    "Electric Range (km)": [random.randint(150, 500) for _ in range(num_records)],
    "Fast Charge Speed (kW)": [random.randint(50, 150) for _ in range(num_records)],
    "Date": [fake.date_this_year() for _ in range(num_records)],
    "Time": [fake.time() for _ in range(num_records)],
    "Month": [fake.month_name() for _ in range(num_records)],
    "Amount for Charging ($)": [round(random.uniform(10, 50), 2) for _ in range(num_records)],
    "Location (Latitude)": [round(fake.latitude(), 6) for _ in range(num_records)],
    "Location (Longitude)": [round(fake.longitude(), 6) for _ in range(num_records)],
    "Speed (km/h)": [random.randint(30, 120) for _ in range(num_records)],
    "Charging Status": [random.choice(["Charging", "Not Charging", "Fully charged"]) for _ in range(num_records)],
    "Service History": [random.choice(["Completed", "Scheduled", "Due Soon"]) for _ in range(num_records)],
    "Alerts/Warnings": [random.choice(["Battery Low", "Service Due", "No Alerts"]) for _ in range(num_records)]
}

# Additional columns for the dataset
extra_columns = {
    "Make": random.choices(["Tesla", "BMW", "Volkswagen", "Polestar"], k=num_records),
    "Manager": random.choices(["Mark", "Vettal", "Sainz", "John", "Mathew"], k=num_records),
    "Acceleration 0 - 100 km/h": [round(random.uniform(3.0, 8.0), 1) for _ in range(num_records)],
    "Top Speed": [random.randint(150, 300) for _ in range(num_records)],
    "ER": [random.randint(400, 600) for _ in range(num_records)],
    "Total Power": [random.randint(150, 400) for _ in range(num_records)],
    "Total Torque": [random.randint(300, 700) for _ in range(num_records)],
    "Drive": random.choices(["AWD", "Rear", "Front"], k=num_records),
    "BC": [random.randint(75, 100) for _ in range(num_records)],
    "Length": [random.randint(4000, 5000) for _ in range(num_records)],
    "Width": [random.randint(1800, 1900) for _ in range(num_records)],
    "Height": [random.randint(1400, 1600) for _ in range(num_records)],
    "Wheelbase": [random.randint(2700, 3000) for _ in range(num_records)],
    "Gross Vehicle Weight (GVWR)": [random.randint(2000, 3000) for _ in range(num_records)],
    "Max. Payload": [random.randint(400, 600) for _ in range(num_records)],
    "Cargo Volume": [random.randint(350, 600) for _ in range(num_records)],
    "Seats": [random.randint(4, 7) for _ in range(num_records)]
}

# Add the additional columns to the data
data.update(extra_columns)

# Convert the combined data to a DataFrame
df = pd.DataFrame(data)

file_path = "synthetic_ev_fleet_extended_dataset.csv"
df.to_csv(file_path, index=False)
print(f"Dataset saved successfully at: {file_path}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for the correlation matrix
numerical_columns = df.select_dtypes(include=["number"]).columns
correlation_matrix = df[numerical_columns].corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
